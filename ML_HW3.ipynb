```python
import six.moves.cPickle as pickle
import gzip
import os
import math
import numpy as np
import numpy.linalg as lin
import scipy.misc
from __future__ import division
import collections
from collections import Counter
```

```python
def dist(a, b):
    w, h  = a.shape
    dist = 0
    for x in range(w):
        for y in range(h):
            dist = dist + (a[x][y] - b[x][y])**2
    return dist
```

```python
def load_data(dataset):
    ''' Loads the dataset

    :type dataset: string
    :param dataset: the path to the dataset (here MNIST)
    
    copied from http://deeplearning.net/ and revised by hchoi
    '''
    # Download the MNIST dataset if it is not present
    data_dir, data_file = os.path.split(dataset)
    if data_dir == "" and not os.path.isfile(dataset):
        # Check if dataset is in the data directory.
        new_path = os.path.join(
            os.path.split(__file__)[0],
            dataset
        )
        if os.path.isfile(new_path) or data_file == 'mnist.pkl.gz':
            dataset = new_path

    if (not os.path.isfile(dataset)) and data_file == 'mnist.pkl.gz':
        from six.moves import urllib
        origin = (
            'http://www.iro.umontreal.ca/~lisa/deep/data/mnist/mnist.pkl.gz'
        )
        print('Downloading data from %s' % origin)
        urllib.request.urlretrieve(origin, dataset)

    print('... loading data')

    # Load the dataset
    with gzip.open(dataset, 'rb') as f:
        try:
            train_set, valid_set, test_set = pickle.load(f, encoding='latin1')
        except:
            train_set, valid_set, test_set = pickle.load(f)
    # train_set, valid_set, test_set format: tuple(input, target)
    # input is a numpy.ndarray of 2 dimensions (a matrix)
    # where each row corresponds to an example. target is a
    # numpy.ndarray of 1 dimension (vector) that has the same length as
    # the number of rows in the input. It should give the target
    # to the example with the same index in the input.

    return train_set, valid_set, test_set
```

```python
train_set, val_set, test_set = load_data('C:/test/mnist.pkl.gz')

train_x, train_y = train_set
val_x, val_y = val_set
test_x, test_y = test_set

print(train_x.shape) # 50k * 784
print(train_y.shape) # 50k

'''
for i in range(100):
    samp_img = train_x[i].reshape((28, 28))
    scipy.misc.imsave('test'+str(i)+'.jpg', samp_img)
    print(train_y[i])
'''        

save_path = "G:/내 드라이브/Class/8th/ML\HW1/"
```

```python
#for saving memory
train_x_sam = train_x[1:501 ]# 500 * 784 
train_y_sam = train_y[1:501] # 500
test_x_sam = test_x[1:101] # 100 * 784
test_y_sam = test_y[1:101] # 100
```

```python
train_x_sam_sq = []
test_x_sam_sq = []
for tr_x in train_x_sam:
    train_x_sam_sq.append(tr_x.reshape(28,28))
for ts_x in test_x_sam:
    test_x_sam_sq.append(ts_x.reshape(28,28))
```

```python
k = 2

predict = []
n=1

for ts_x in test_x_sam_sq:
    
    dist_list = []
    ts_y = list(test_y_sam)
    tr_y = list(train_y_sam)
    nearest_k = []
    
    # distance with 
    for tr_x in train_x_sam_sq:
        dist_list.append(dist(tr_x, ts_x))
        
    for find_max in range(k):
        index = dist_list.index(max(dist_list))
        x_pop = dist_list.pop(index)
        y_pop = tr_y.pop(index)
        nearest_k.append(y_pop)
    
    counts = Counter(nearest_k)
    max_cnt = max(Counter(nearest_k).values())
    for key in counts.keys():
        if(counts[key] == max_cnt):
            break
    predict.append(key)
    
    print(str(n)+"/"+str(len(test_x_sam)))
    n = n+1
```
